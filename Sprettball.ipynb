```python
# importerer to biblioteker og gir dem aliaser
import numpy as np 
import matplotlib.pyplot as plt

# Leser av data og lagrer det som en 2D array
data = np.loadtxt("https://raw.githubusercontent.com/LektorB/Fysikk-1/main/ball2.txt",
                  skiprows = 2,delimiter = ',',dtype='str',encoding = 'utf-8' )
data = np.char.replace(data, '−', '-').astype('float64')


tid = data[:,0]
pos = data[:,1]

# Lager og pynter graf
plt.plot(tid,pos)
plt.title('Posisjonsgraf - Sprettball')
plt.xlabel('$t$/s')
plt.ylabel('$s$/m')
plt.show()
```

pos


```python
n = len(tid)
vel = np.zeros(n-1)

for i in range(1,n-1):
    delta_s = pos[i+1] - pos[i-1]
    delta_t = tid[i+1] - tid[i-1]
    fart = delta_s/delta_t
    vel[i] = fart

plt.plot(tid[:-1],vel)
plt.title('Fartssgraf - Sprettball')
plt.xlabel('$t$/s')
plt.ylabel('$v$/(m/s)')
plt.show()
```

```python
aks = np.zeros(n-2)

for i in range(1,n-2):
    delta_v = vel[i+1] - vel[i-1]
    delta_t = tid[i+1] - tid[i-1]
    akselerasjon = delta_v/delta_t
    aks[i] = akselerasjon

plt.plot(tid[:-2],aks)
plt.title('Akselerasjonsgraf - Sprettball')
plt.xlabel('$t$/s')
plt.ylabel('$a$/(m/s$^2$)')
plt.show()
```
