```python
# importerer to biblioteker og gir dem aliaser
import numpy as np 
import matplotlib.pyplot as plt

# Leser av data og lagrer det som en 2D array
data = np.loadtxt("https://raw.githubusercontent.com/LektorB/Fysikk-1/main/ball2.txt",
                  skiprows = 2,delimiter = ',',dtype='str',encoding = 'utf-8' )
data = np.char.replace(data, '−', '-').astype('float64')


tid = data[:,0]
pos = data[:,1]

# Lager og pynter graf
plt.plot(tid,pos)
plt.title('Posisjonsgraf - Sprettball')
plt.xlabel('$t$/s')
plt.ylabel('$s$/m')
plt.show()
```
